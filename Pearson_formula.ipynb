# Pearson_formula

```python
x = [1, 2, 3, 4, 5, 6]
y = [7, 8, 9, 16, 12, 11]



## xy
xy = []
for num1, num2 in zip(x, y):
    xy.append(num1 * num2)
print(xy)


def summation(series):
    total = 0
    for i in series:
        total+=i
    return total


def square(series):
  return [i ** 2 for i in series]


x_square=square(x)
y_square=square(y)


x_sum=print(summation(x))
y_sum=print(summation(y))


print(f'x*y:::{summation(xy)}')


print(f' x square summation  :: {summation(x_square)}')
print(f' y square summation  :: {summation(y_square)}')

```

```python
mylist=[]

def difference(series):
    mylist.clear()
    sum=0
    
    num=len(series)

    for i in series:
        sum+=i
    avg=sum / num
    for i in series:
        r=i-avg
        mylist.append(r)

    return mylist
```

```python
li=[]
def difference_y(series):
    mylist.clear()
    sum=0
    
    num=len(series)

    for i in series:
        sum+=i
    avg=sum / num
    for i in series:
        r=i-avg
        li.append(r)

    return li
y_b=difference_y(y)
```

```python
xBar=difference(x)
xBar
```

```python
products = []


for num1, num2 in zip(xBar, y_b):
	products.append(num1 * num2)
formula_part1=summation(products)
formula_part1
```
